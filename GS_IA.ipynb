```python
# Função para verificar se a distribuição é normal
def verifica_normalidade(dataframe, coluna):
    coluna_data = dataframe[coluna]
    # Realizar o teste de Shapiro-Wilk
    statistic, p_valor = shapiro(coluna_data)
    # Definir o nível de significância
    nivel_significancia = 0.1
    # Verificar se a hipótese nula de normalidade pode ser rejeitada
    if p_valor > nivel_significancia:
        return True
    else:
        return False
```

# **Explicação das Configurações**
hidden_layer_sizes=(5, 4):

**Número de Camadas Ocultas**: A rede tem duas camadas ocultas.
Neurônios por Camada: A primeira camada oculta tem 5 neurônios e a segunda camada oculta tem 4 neurônios.
activation='relu':

**Função de Ativação**: relu (Rectified Linear Unit) é usada como a função de ativação. Esta é uma escolha comum porque relu ajuda a evitar problemas de gradientes desaparecendo e geralmente permite um treinamento mais rápido e eficaz em redes profundas.
learning_rate_init=0.00001:

**Taxa de Aprendizado Inicial**: Define a taxa de aprendizado inicial para o otimizador. Um valor de 0.00001 é bastante pequeno, o que pode ajudar a garantir que o modelo converge lentamente e de forma estável, especialmente útil se os dados forem complexos e sensíveis a grandes atualizações de peso.
max_iter=500:

**Número Máximo de Iterações**: Define o número máximo de iterações do otimizador. Com 500 iterações, há um balanço entre dar tempo suficiente para a rede aprender os padrões dos dados e evitar overfitting.
tol=0.000001:

**Tolerância para o Critério de Parada**: Se a melhoria na função objetivo (erro de treinamento) entre duas iterações consecutivas for menor que 0.000001, o treinamento será interrompido. Um valor tão pequeno significa que estamos buscando uma convergência muito precisa.
verbose=False:

**Verbosidade**: Define se as mensagens de progresso devem ser exibidas durante o treinamento. Com verbose=False, essas mensagens são suprimidas para manter a saída limpa.

# **Justificativa para a Estrutura de Camadas**
1. Escolha das Camadas Ocultas (5, 4):
Primeira Camada Oculta com 5 Neurônios:
Equilíbrio entre Complexidade e Capacidade de Generalização: Ter 5 neurônios permite à rede capturar padrões não-lineares nos dados sem introduzir complexidade excessiva. Um número pequeno de neurônios ajuda a prevenir overfitting e reduz a carga computacional.
Segunda Camada Oculta com 4 Neurônios:
Refinamento de Padrões: A segunda camada com 4 neurônios permite à rede refinar os padrões aprendidos pela primeira camada, capturando interações mais complexas entre as características dos dados.
2. Função de Ativação relu:
Vantagens do relu: relu é popular em redes neurais profundas devido à sua capacidade de mitigar problemas de gradientes desaparecendo, permitindo que a rede aprenda mais rapidamente e eficientemente. Também é computacionalmente eficiente.
3. Taxa de Aprendizado Inicial Baixa (0.00001):
Convergência Estável: Uma taxa de aprendizado baixa garante que as atualizações dos pesos durante o treinamento sejam pequenas, promovendo uma convergência estável. Isso é particularmente útil quando se ajusta um modelo em dados complexos ou sensíveis a grandes mudanças.
4. Número Máximo de Iterações (500):
Treinamento Suficiente: Com 500 iterações, o modelo tem tempo suficiente para ajustar seus pesos, especialmente importante quando a taxa de aprendizado é baixa.
5. Critério de Parada Rigoroso (tol=0.000001):
Busca por Precisão: Uma tolerância baixa exige que o modelo alcance uma melhoria muito pequena na função de perda entre iterações consecutivas antes de parar, ajudando a garantir uma boa convergência.

# **Por que usar k=1?**
Simplicidade: Usar k=1 é a forma mais simples de KNN, onde cada ponto é classificado com base apenas no seu vizinho mais próximo.
Evita Misturas: Com k=1, a classificação é direta e não envolve a combinação de votos de múltiplos vizinhos, o que pode ser útil se os pontos mais próximos forem bons representantes das classes.

Ajuste Fino: Em algumas situações, especialmente em conjuntos de dados com poucas amostras ou quando os dados são bem separados, k=1 pode funcionar bem.

# **Considerações sobre k=1**
Overfitting: O uso de k=1 pode levar a overfitting, onde o modelo se ajusta muito aos dados de treino, mas não generaliza bem para novos dados. Isso ocorre porque o modelo se torna muito sensível a ruídos nos dados.
Dependência de Dados: A eficácia de k=1 depende fortemente da distribuição dos dados. Em conjuntos de dados com muito ruído ou sobreposição entre as classes, k=1 pode não ser ideal.

Avaliação de k Ideal: Normalmente, é melhor experimentar com diferentes valores de k para encontrar o que funciona melhor para o seu conjunto de dados específico. Isso pode ser feito usando validação cruzada, como mostrado anteriormente.
